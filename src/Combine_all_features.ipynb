# Load individual pickle files

## Loading User features

```python
df_tweet_features_path = "/content/drive/MyDrive/Twibot-20/final_dataframes/tweet_stats_pd.pickle"
df_user_features_path = '/content/drive/MyDrive/Twibot-20/final_dataframes/user_features.pickle'

df_tweet_features = pd.read_pickle(df_tweet_features_path)
df_user_features = pd.read_pickle(df_user_features_path)
```

```python
df_user_features
```

```python
df_tweet_features = pd.DataFrame.from_dict(df_tweet_features, orient='index')
final_features_df = df_user_features.merge(df_tweet_features, left_index=True, right_index=True, how='inner')
```

```python
final_features_df
```

```python
final_features_df = final_features_df.drop(columns=['Links_M2', 'Words_M2', 'Hashtags_M2', "Mentions_M2"])
```

```python
final_features_df
```

```python
final_features_df['RT'] = final_features_df.apply(lambda row: row['RT']/row['Number'], axis=1)
```

```python
final_features_df
```

```python
from scipy.stats import zscore

final_features_df[
    ['Number', 'public_metrics.followers_count', 'public_metrics.following_count', 'public_metrics.tweet_count', 'public_metrics.listed_count']] = final_features_df[['Number', 'public_metrics.followers_count', 'public_metrics.following_count', 'public_metrics.tweet_count', 'public_metrics.listed_count']].apply(zscore)
```

```python
final_features_df
```

```python
import pickle
with open('df_final_features16.pickle', 'wb') as f:
    # Use the pickle.dump() function to save the dictionary to the file
    pickle.dump(final_features_df, f)
```
